```python
Measure of Central tendency
Mean: Average (Mu)
Median: Middle number of a dataset. In case the dataset has even rows, average the 2 center numbers.
Mode: Value that occurs most often.
```

```python
Measure of Asymmetry
Skewness: indicates if the data is concentrated on one side. If data is concentracted on the left, it is
   called right skewed asoutliers are to the right (mean > median) and the other way. if equally distributed then the data is asymmentrical.
```

```python
Measure of Variability
Variance: measures the dispersoin of the set of datapoints around the mean. Sigma^2 = Sum (Value - Mean)^2 / Number of Observations
      .this is population formula that is to be applied to complete dataset.
     When only a sample of data is used then, Sample formula(S for sample) S^2 = Sum (Xi - Mu)^2 /(N-1)
Standard Deviation:  Sigma = Sqrt of Sigma^2
Coefficient of Variaance (cv):  cv = Sigma/Mean  it is also called relative standard deviation. This is helpful in compareing seperate datasets
      whwre the range of data is different. cv will always range form -1 to 1
```

```python
determine the relation between variables
Covariance:    Cov= sum of ((Var1 - Mean1)*(Var2-Mean2))/(N observations -1)   --> -1 for sample formula only
      the covariance has positive number if the relation between 2 varialbles move in the same direction. eg when area increas the price also increase.
      if covariance is negative for opposite relation between variables. like odometer vs Price
      0 if independent.
      value can be too low or too high.
Linear correlation Coefficient:  is Cov(x,y) /(Standard deviation(x) * StDev(y))
      it will always between -1 to 1
```
